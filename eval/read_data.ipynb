```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
```

```python
def path_to_data(experiment, example=True, mu=0):
    if example:
        data_path = "../example_data/"
        data_path += experiment +".npy"
    else:
        data_path = "../data/"+experiment+"/" + experiment + "_{}.jld2".format(mu)
    return data_path
```

```python
data = np.load(path_to_data("lorenz"))
data.shape
```

```python
axis = plt.subplot(projection = "3d")
axis.plot(data[:,0],data[:,1],data[:,2])
```

```python
f = h5py.File(path_to_data("lorenz", example=False, mu=28))
list(f.keys())
tseries = f["time series"][:,:]
data = tseries.T
```

```python
axis = plt.subplot(projection = "3d")
axis.plot(data[:,0],data[:,1],data[:,2])
```
