# check the data


```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
```

```python
def path_to_data(experiment, example=True, mu=0):
    if example:
        data_path = "../example_data/"
        data_path += experiment +".npy"
    else:
        data_path = "../experiments/data/"+experiment+"/" + experiment + "_{}.jld2".format(mu)
    return data_path
```

```python
data_npy = np.load(path_to_data("bursting_neuron"))
data_npy.shape
t = 10000
```

```python
axis = plt.subplot(projection = "3d")
axis.plot(data_npy[:t,0],data_npy[:t,1],data_npy[:t,2])
```

```python
f = h5py.File(path_to_data("bursting_neuron", example=False, mu=10.2))
list(f.keys())
tseries = f["time series"][:,:]
data = tseries.T
```

```python
axis = plt.subplot(projection = "3d")
axis.plot(data[:t,0],data[:t,1],data[:t,2])
```
