# check the data


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def path_to_data(experiment, example=True):
    if example:
        data_path = "../example_data/"
    else:
        data_path = "../data/"
    return data_path + experiment + ".npy"
```

```python
data = np.load(path_to_data("bursting_neuron"))
data.shape
```

```python
axis = plt.subplot(projection = "3d")
axis.plot(data[:,0],data[:,1],data[:,2])
```
